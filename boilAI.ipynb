```python
"""
@author: EJSU
"""
"""
implement factbase using dictionary
"""


"""
/*rules*/

R1: IF 
water level is high and
temperature is high and
the release is closed or 
(the release is not closed and flow rate is high)
HEN Replace the outlet pipe
R2: IF 
water level is low and
temperature is high and
the release is not closed 
THEN open control valve and shut down tubes
R3: IF
water level is high and
temperature is low 
THEN no action needed
R4: IF
water level is high and
temperature is high and
(the release is not closed and flow rate is low)
THEN no action needed
R5: IF
water level is low and
temperature is low 
THEN open control valve
"""

q = { 
        'water level' : False,
        'temperature' : False,
        'release' : False,
        'flow rate' : False,
        }


new_fact = True


"""
Implement rules fired using list
"""

rules_fired = ['facts']

while new_fact == True: 
    
    # new cycle
    new_fact = False  

    val = "begin"
    print("Please answer the following question.\n")
    print("What is the transducer output?(water level)\n")
    val = input()

    #q1    
    if val == "high":
      q['water level'] = True
    else :
      q['water level'] = False

    #q2
    print("What is the temperature?\n")  
    val = input()
    
    if val=="high":
      q['temperature'] = True
    else :
      q['temperature'] = False

    #results: no action needed or open control valve

    if q['water level'] is True and q['temperature'] is False and 'R3' not in rules_fired :
      rules_fired.append('R3')
      print('No action needed!\n') 
      new_fact = True
      break
    elif q['water level'] is False and q['temperature'] is False and 'R5' not in rules_fired :
      rules_fired.append('R5')
      print('You should open control valve!\n')
      new_fact = True
      break
    else:
      pass

    #q3
    print("Is the release valve closed?\n")  
    val = input()
    
    if val=="closed":
      q['release'] = True
    else :
      q['release'] = False

    #result: open control valve and shut down tubes

    if q['water level'] is False and q['temperature'] is True and q['release'] is False and 'R2' not in rules_fired :
      rules_fired.append('R2')
      print('You should open control valve and shut down tubes!\n')
      new_fact = True
      break
    else:
      pass

   #q4
    print("What is the flow rate?\n")  
    val = input()
    
    if val=="high":
      q['flow rate'] = True
    else :
      q['flow rate'] = False


    #result: Replace the outlet pipe or no action needed

    if q['water level'] is True and q['temperature'] is True and q['flow rate'] is True and 'R1' not in rules_fired  :
      rules_fired.append('R1')
      print('You should replace the outlet pipe!\n')
      new_fact = True
      break
    elif q['water level'] is True and q['temperature'] is True and q['flow rate'] is False and q['flow rate'] is False and 'R4' not in rules_fired :
      rules_fired.append('R4')
      print('No action needed!\n')
      new_fact = True
      break

```
